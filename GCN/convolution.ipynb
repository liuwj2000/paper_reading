# 卷积

这个特征有什么意义？

我们令x=t，y=n-t，那么x+y=n 就是下面这些直线：

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-2,2,5)
y1=-2-x
y2=-1-x
y3=-x
y4=1-x
y5=2-x
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.plot(x,y5)
plt.show()
```

这样就好像一块毛巾沿着角卷起来一样
